```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os.path
import os
from os.path import join
from threading import Thread
import time
import gc

from scipy import misc
```

```python
input_folders = ['predictions_third_best', 'predictions_fourth_best', 'predictions_fifth_best']
save_folder = 'comb_3_4_5'
number_of_threads = 100
img_list = os.listdir(input_folders[0])
```

```python
def work_img(fname, input_folders, save_folder):
    img1_1 = misc.imread(join(input_folders[0], fname)).astype(int)
    img2_1 = misc.imread(join(input_folders[1], fname)).astype(int)
    img3_1 = misc.imread(join(input_folders[2], fname)).astype(int)
    
    img = (img1_1 + img2_1 + img3_1)/3
    img[img<127] = 0
    img[img>=127] = 1
    misc.imsave(join(save_folder, fname), img)
```

```python
cnt = 0
while(len(img_list) != 0):
    threads = []
    for i in range(number_of_threads):
        fname = img_list.pop(0)
        thread = Thread(target=work_img, args=(fname, input_folders, save_folder))
        thread.start()
        threads.append(thread);
        if(len(img_list) == 0):
            break
    
    for t in threads:
        t.join()
        
    cnt += number_of_threads
    print(str(cnt) + ' done')
```
